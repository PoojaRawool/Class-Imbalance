# Random under sampling

### Separate majority and minority classes

```python
data["class"].value_counts()
```

```python
data_majority=data[data["class"]=="n"]
data_minority=data[data["class"]=="w"]
```

### Downsample majority class

```python
from sklearn.utils import resample
data_majority_downsampled=resample(data_majority,replace=False,n_samples=74,random_state=123)
```

### Combine minority class with downsampled majority class

```python
data_downsampled=pd.concat([data_majority_downsampled,data_minority])
```

### Display new class count

```python
data_downsampled["class"].value_counts()
```

```python
data_downsampled.shape
```

As you can see new dataframe has fewer observations than the original and the ratio of 2 classes is now 1:1.

```python
y=data_downsampled["class"]
x=data_downsampled.drop("class",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear',C=10,random_state=0).fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
clf.score(x_test, y_test)
```

With down sampling accuracy has reduced to 96%

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test,y_pred)
print(conf_mat)
```

1 tuple was having class "n" as its actual class and was missclassified as class "w" 

Only one tuple with class "n" were missclassified.

As tuples with class "w" are less i.e. 74. I am considering class "w" as the class of my interest. Random under sampling is giving good results.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

Precision:<br>
out of all tuples with class "n" prediction 94% I got it right.<br>
out of all tuples with class "w" prediction 100% I got it right.<br>
Recall:<br>
out of all tuples with class "n" truth 100% I got it right.<br>
out of all tuples with class "w" truth 93% I got it right.<br>

```python
from sklearn.metrics import roc_auc_score
prob=clf.predict_proba(x_test)
```

```python
prob
```

```python
# keep only positive class
prob=[p[1] for p in prob]
```

```python
prob
```

```python
print(roc_auc_score(y_test,prob))
```

With respect to the original model trained on the imbalanced dataset got the AUROC score of 98% and when the dataset was down sampled got the AUROC score of 100%.

# Random over sampling

### Upsample minority class

```python
from sklearn.utils import resample
data_minority_upsampled=resample(data_minority,replace=True,n_samples=4265,random_state=123)
```

### Combine majority class with upsampled minority class

```python
data_upsampled=pd.concat([data_minority_upsampled,data_majority])
```

```python
data_upsampled.shape
```

As you can see new dataframe has more observations than the original and the ratio of 2 classes is now 1:1.

### Display new class count

```python
data_upsampled["class"].value_counts()
```

```python
y=data_upsampled["class"]
x=data_upsampled.drop("class",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear',C=10,random_state=0).fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
clf.score(x_test, y_test)
```

With up sampling the accuracy seems to be higher as compared to the down sampling accuracy.

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test,y_pred)
print(conf_mat)
```

# SMOTE

```python
y=data["class"]
x=data.drop("class",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
x_train_smote, y_train_smote = oversample.fit_resample(x_train, y_train)
```

```python
x_train_smote1=x_train_smote.copy()
y_train_smote1=y_train_smote.copy()
```

```python
df_smote=x_train_smote1
```

```python
df_smote["class"]=y_train_smote1
```

```python
df_smote["class"].value_counts()
```

```python
df_smote.shape
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear',C=10,random_state=0).fit(x_train_smote,y_train_smote)
```

```python
y_pred = clf.predict(x_test)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test,y_pred)
print(conf_mat)
```
